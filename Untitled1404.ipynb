```python
#Par ou ímpar

# entrada
print("Verificador de Par ou Ímpar")
numero_par_impar = int(input("Digite um número inteiro: "))

if numero_par_impar % 2 == 0:
    print(f"O número {numero_par_impar} é PAR.")
else:
    print(f"O número {numero_par_impar} é ÍMPAR.")
```

```python
#Maior de dois números

# entrada
numero1 = float(input("Digite o primeiro número: "))
numero2 = float(input("Digite o segundo número: "))

if numero1 > numero2:
    print(f"O primeiro número ({numero1}) é o maior.")
elif numero2 > numero1:
    print(f"O segundo número ({numero2}) é o maior.")
else:
    print("Os dois números são iguais.")
```

```python
#Aprovação de aluno

# entrada
print("Aprovação de Aluno")
nota = float(input("Digite a nota do aluno: "))

# saida

if nota >= 7:
    print("Situação: Aprovado")
elif nota >= 5 and nota < 7:
    print("Situação: Recuperação")
else:
    print("Situação: Reprovado")
```

```python
#Classificação de idade

#entrada
idade = int (input ("digite sua idade: "))

# processamento/saida

if idade <= 12:
   print (f"voce tem {idade} anos de idade, voce é uma criança.")

elif idade <=17:
  print (f"voce tem {idade} anos de idade, voce é adolescente.")

elif idade <=59:
  print (f"voce tem {idade} anos de idade, voce é adulto.")

else:
    print (f"voce tem {idade} anos de idade, voce é idoso.")
```

```python
#Calculadora simples

# Entrada

a= int(input('Digite um número: '))
b= int(input('Digite um número: '))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b

#saída
print ("")
print("--------------------------OPERADORES ARITIMÉTICOS--------------------")
print("")
print("----------------------------------SOMA-------------------------------")
print("")
print (f"O resultado da SOMA de {a} mais {b} é igual a : {soma}")
print("")
print("----------------------------------SUBTRAÇÃO---------------------------")
print("")
print (f"O resultado da SUBTRAÇÃO de {a} menos {b} é igual a : {subtracao}")
print("")
print("---------------------------multiplicacao-----------------------------")
print("")
print(f"o resultado da multiplicacao de {a} vezes {b} é igual a : {multiplicacao}")
print("---------------------------divisao-----------------------------------")
print("")
print(f"o resultado da divisao de {a} dividido {b} é igual a : {divisao}")
```

```python
#Dia da semana

#entrada
dia_numero = int(input("Digite o número do dia"))

#processamento/saida
if dia_numero == 1:
    dia_nome = "domingo"

elif dia_numero == 2:
    dia_nome = "segunda-feira"

elif dia_numero == 3:
    dia_nome = "terça-feira"

elif dia_numero == 4:
    dia_nome = "quarta-feira"

elif dia_numero == 5:
    dia_nome = "quinta-feira"

elif dia_numero == 6:
    dia_nome = "sexta-feira"

elif dia_numero == 7:
    dia_nome = "sábado"

else:
    dia_nome = "Número de dia inválido"

print(f"O dia {dia_numero} corresponde a {dia_nome}.")
```

```python
#Ano bissexto

# entrada
print("Verificador de Ano Bissexto")
ano = int(input("Digite um ano: "))

# saida
if (ano % 4 == 0 and ano % 100 != 0) or (ano % 400 == 0):
    print(f"O ano {ano} é um ano bissexto.")
else:
    print(f"O ano {ano} não é um ano bissexto.")
```

```python
#Desconto em compras

# entrada
print("Calculadora de Desconto")
valor_compra = float(input("Digite o valor da compra: "))

#saida
desconto = 0
if valor_compra >= 100:
    desconto = 0.10 # 10%
elif valor_compra >= 50:
    desconto = 0.05 # 5%
else:
    desconto = 0 # Sem desconto

valor_final = valor_compra * (1 - desconto)

print(f"Valor da compra: R$ {valor_compra:.2f}")
print(f"Desconto aplicado: {desconto * 100:.0f}%")
print(f"Valor final com desconto: R$ {valor_final:.2f}")
```
