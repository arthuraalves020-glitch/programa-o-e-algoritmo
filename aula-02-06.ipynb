```python
#fazer um jogo de adivinhação

# o usuario ira digitar um número de 0 a 10 para adivinhar o número secreto
# o usuario tera 03 tentativas para acertar o número
# o programa devera imprimir uma mensagem se o paulpite do usuario e maior ou menor que o número secreto
# o programa devera imprimir uma mnesagem de quente, morno ou frio

# regras

# se a diferença do paulpite para o número secreto for 3- frio, 2- morno e 1 quente

import random

numero_secreto = random.randint(0, 10)
tentativas = 0
maximastentativas = 3

print("Bem vindo ao jogo de adivinhação!")
print(f"Estou pensando em um número entre 0 e 10. Você tem {maximastentativas} tentativas.")

while tentativas < maximastentativas:
    try:
        guess = int(input("Digite seu palpite (entre 0 e 10): "))
        if not (0 <= guess <= 10):
            print("Por favor, digite um número entre 0 e 10.")
            continue

        tentativas += 1

        if guess == numero_secreto:
            print(f"Parabéns! Você adivinhou o número secreto {numero_secreto} em {tentativas} tentativas!")
            break
        else:
            print("Palpite incorreto.")
            if guess < numero_secreto:
                print("O número secreto é MAIOR que o seu palpite.")
            else:
                print("O número secreto é MENOR que o seu palpite.")

            difference = abs(numero_secreto - guess)
            if difference == 1:
                print("Você está QUENTE!")
            elif difference == 2:
                print("Você está MORNO.")
            elif difference >= 3:
                print("Você está FRIO.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if tentativas == maximastentativas and guess != numero_secreto:
    print(f"Suas {maximastentativas} tentativas acabaram. O número secreto era {numero_secreto}.")
```
