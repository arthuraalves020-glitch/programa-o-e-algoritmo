```python
while True:
    try:
        num = int(input("Digite um número inteiro positivo: "))
        if num > 0:
            break
        else:
            print("Por favor, digite um número positivo.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

soma = 0
contador = 1
while contador <= num:
    soma += contador
    contador += 1
print(f"Resultado: {soma}")

print("\n" * 2) # Add some space for readability
```

```python
while True:
    try:
        num_tabuada = int(input("Digite um número para ver sua tabuada: "))
        break
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

contador_tabuada = 1
while contador_tabuada <= 10:
    resultado_tabuada = num_tabuada * contador_tabuada
    print(f"{num_tabuada} x {contador_tabuada} = {resultado_tabuada}")
    contador_tabuada += 1

print("\n" * 2) # Add some space for readability
```

```python
while True:
    try:
        num_fatorial = int(input("Digite um número inteiro positivo para calcular o fatorial: "))
        if num_fatorial >= 0:
            break
        else:
            print("Por favor, digite um número positivo ou zero.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

if num_fatorial == 0:
    print("Fatorial = 1")
elif num_fatorial < 0:
    # This case should be handled by the while loop, but as a fallback
    print("Fatorial não definido para números negativos.")
else:
    fatorial = 1
    i = 1
    while i <= num_fatorial:
        fatorial *= i
        i += 1
    print(f"Fatorial = {fatorial}")

print("\n" * 2) # Add some space for readability
```

```python
notas = [100, 50, 20, 10, 5, 2, 1]

while True:
    try:
        saque = int(input("Digite o valor do saque: "))
        if saque > 0:
            break
        else:
            print("Por favor, digite um valor de saque positivo.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número inteiro.")

valor_restante = saque

i = 0
while i < len(notas):
    nota = notas[i]
    quantidade_notas = 0
    while valor_restante >= nota:
        valor_restante -= nota
        quantidade_notas += 1

    if quantidade_notas > 0:
        print(f"{quantidade_notas} nota(s) de R${nota}")

    i += 1
```
