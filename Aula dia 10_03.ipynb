Escrever um programa que calcula a média das notas do aluno.
entrada:nome, idade, e as 04 notas do bimestre
processamento: calcular a média
saída: mostrar o nome do aluno e a média obtida.

```python
from typing_extensions import final
from re import X
#entrada
nome = input("digite seu nome:")
idade = int(input("digite sua idade:"))
nota1 = int(input("digite sua nota:"))
nota2 = int(input("digite sua nota:"))
nota3 = int(input("digite sua nota:"))
nota4 = int(input("digite sua nota:"))


#processamento
nota = (nota1 + nota2 + nota3 + nota4) / 4

#saída
print(f"olá {nome}, seja bem vindo!")
print (f"você tem {idade} anos.")
print(f"sua média é {nota}")
```
