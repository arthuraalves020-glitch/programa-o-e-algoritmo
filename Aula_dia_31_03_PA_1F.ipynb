```python
#estrutura de decisões


#escrever um programa que recebe o nome e a idade do usuario e verifica se o usuario e criança, adolescente e adulto
# >= anos e criança
# <= 17 anos e adolescente
# >= 18 anos e adulto

#entrada

nome = str (input ("digite o seu nome: "))
idade = int (input ("digite sua idade: "))

# processamento/saida

if idade <= 12:
   print (f"{nome}, voce tem {idade} anos de idade, voce é criança.")

elif idade <=17:
  print (f"{nome}, voce tem {idade} anos de idade, voce é adolescente.")

else:
    print (f"{nome}, voce tem {idade} anos de idade, voce é adulto.")
```

```python
#em um jogo, um personagem possui uma variavel chamada energia. o programa deve:
# se a energia for maior que 70 - Forte, se a energia estiver entre 30 e 70 - cansado,
#se a energia estiver menor que 30 -fraco

#entrada
energia = int(input("digite o nível de energia do personagem: "))

#processamento/saida
if energia <= 30:
    print(f"seu personagem tem energia igual a {energia}. ele esta fraco.")
elif energia <= 70:
    print(f"seu personagem tem energia igual a {energia}. ele esta cansado.")
else:
    print(f"seu personagem tem energia igual a {energia}. ele esta Forte.")
```

```python
#escrever um programa que classifica os tipos de triangulo e valida se e um triangulo.

#validar se os lados digitados para formar um triangulo
# lado 1 + lado2 > lado3 and lado1 + lado3 > lado 2 and lado2 + lado3 > lado1

#tipos de triangulo
#equilatero - todos os lados iguais - ==
#escaleno - todos os lados diferentes - !=
#isóceles - dois lados iguais

#entrada

lado1 = int(input("digite o primeiro lado do triangulo: "))
lado2 = int(input("digite o segundo lado do triangulo: "))
lado3 = int(input("digite o terceiro lado do triangulo: "))

#verifica se e um triangulo valido
if lado1 + lado2 > lado3 and lado1 + lado3 > lado2 and lado2 + lado3 > lado1:
    #saida - classifica o tipo de triangulo
    if lado1 == lado2 == lado3:
        print(f"Os lados {lado1}, {lado2}, {lado3} formam um triangulo equilatero.")
    elif lado1 != lado2 and lado1 != lado3 and lado2 != lado3:
        print(f"Os lados {lado1}, {lado2}, {lado3} formam um triangulo escaleno.")
    else:
        print(f"Os lados {lado1}, {lado2}, {lado3} formam um triangulo isóceles.")
else:
    print(f"Os lados {lado1}, {lado2}, {lado3} não formam um triangulo valido.")
```
