```python
saldo = 1000

while True:
    print("\n--- CAIXA ELETRÔNICO ---")
    print("1. Ver Saldo")
    print("2. Depositar")
    print("3. Sacar")
    print("4. Sair")

    opcao = input("Escolha uma opção: ")

    if opcao == '1':
        print(f"Seu saldo atual é: R${saldo:.2f}")
    elif opcao == '2':
        valor_deposito = float(input("Digite o valor a depositar: "))
        if valor_deposito > 0:
            saldo += valor_deposito
            print(f"Depósito de R${valor_deposito:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")
        else:
            print("Valor de depósito inválido.")
    elif opcao == '3':
        valor_saque = float(input("Digite o valor a sacar: "))
        if valor_saque > 0 and valor_saque <= saldo:
            saldo -= valor_saque
            print(f"Saque de R${valor_saque:.2f} realizado com sucesso. Novo saldo: R${saldo:.2f}")
        elif valor_saque > saldo:
            print("Saldo insuficiente.")
        else:
            print("Valor de saque inválido.")
    elif opcao == '4':
        print("Obrigado por usar o Caixa Eletrônico! Volte sempre!")
        break
    else:
        print("Opção inválida. Por favor, escolha uma opção de 1 a 4.")
```
