```python
# Operadores aritiméticos
# Fazer um programa que calcula operações com 02 números

# Entrada

a= int(input('Digite um número: '))
b= int(input('Digite um número: '))

# Processamento

soma = a + b
subtracao = a - b
multiplicacao = a * b
divisao = a / b
div_int = a // b
resto = a % b
potencia = a ** b
#saída
print ("")
print("--------------------------OPERADORES ARITIMÉTICOS--------------------")
print("")
print("----------------------------------SOMA-------------------------------")
print("")
print (f"O resultado da SOMA de {a} mais {b} é igual a : {soma}")
print("")
print("----------------------------------SUBTRAÇÃO---------------------------")
print("")
print (f"O resultado da SUBTRAÇÃO de {a} menos {b} é igual a : {subtracao}")
print("")
print("---------------------------multiplicacao-----------------------------")
print("")
print(f"o resultado da multiplicacao de {a} vezes {b} é igual a : {multiplicacao}")
print("---------------------------divisao-----------------------------------")
print("")
print(f"o resultado da divisao de {a} dividido {b} é igual a : {divisao}")
print("")
print("---------------------------divisao inteira---------------------------")
print("")
print(f"o resultado da divisao inteira de {a} dividido {b} é igual a : {div_int}")
print("")
print("---------------------------resto da divisao---------------------------")
print("")
print(f"o resultado da divisao de {a} dividido {b} é igual a : {resto}")
print("")
print("---------------------------potencia----------------------------------")
print("")
print(f"o resultado da potencia de {a} elevado a {b} é igual a : {potencia}")
print("")
print("---------------------------------------------------------------------")
```

```python
#operadores relacionais


#entrada

a= int(input('Digite um número: '))
b= int(input('Digite um número: '))

# Processamento

#comparacao

maior = a > b
menor = a < b
maior_igual = a >= b
menor_igual = a <= b

# saída
print(maior)
print(menor)
print(maior_igual)
print(menor_igual)
```
