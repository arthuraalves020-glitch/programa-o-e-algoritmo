```python
#Escrever um programa que o usuário digita um número de 1 a 10 e o programa impri a tabuada do número digitado

numero = int(input("Digite um número de 1 a 10 para ver a tabuada: "))

if 1 <= numero <= 10:
    print(f"Tabuada do {numero}:")
    for i in range(1, 11):
        print(f"{numero} x {i} = {numero * i}")
elif numero == 0:
    print("Não há tabuada para 0")
else:
    print("Número fora do intervalo de 1 a 10.")
```

```python
# 1 - Escrever um programa que o usuário digita um número e o programa imprimi até o número digitado.
numero = int(input("Digite um número: "))

if numero < 1:
    print("Por favor, digite um número positivo.")
else:
    print(f"Números de 1 até {numero}:")
    for i in range(1, numero + 1):
        print(i)
```

```python
# 2 - Escrever um programa que o usuário digita um número e o programa imprimi o número digitado até o zero (contagem regressiva).numero = int(input("Digite um número para a contagem regressiva: "))

if numero < 0:
    print("Por favor, digite um número não negativo.")
else:
    print(f"Contagem regressiva a partir de {numero}:")
    for i in range(numero, -1, -1):
        print(i)
```

```python
# 3 - Escrever um programa que o usuário digita um número e o programa imprimi os números pares.numero = int(input("Digite um número para ver os pares até ele: "))

if numero < 1:
    print("Por favor, digite um número positivo.")
elif numero == 1:
    print("Não há números pares até 1.")
else:
    print(f"Números pares de 1 até {numero}:")
    for i in range(1, numero + 1):
        if i % 2 == 0:
            print(i)
```

```python
# 4 - Escrever um programa que o usuário digita um número e o programa imprimi os números impares.
numero = int(input("Digite um número para ver os ímpares até ele: "))

if numero < 1:
    print("Por favor, digite um número positivo.")
else:
    print(f"Números ímpares de 1 até {numero}:")
    for i in range(1, numero + 1):
        if i % 2 != 0:
            print(i)
```

```python
# 5 - Escrever um programa que o usuário digita um número e programa soma os números até o número digitado.numero = int(input("Digite um número para somar até ele: "))

soma = 0
if numero < 1:
    print("Por favor, digite um número positivo.")
else:
    for i in range(1, numero + 1):
        soma += i
    print(f"A soma dos números de 1 até {numero} é: {soma}")
```

```python
# 6 - Escrever um programa que o usuário digita uma palavra e o programa conta as letras da palavra.
palavra = input("Digite uma palavra para contar suas letras: ")

numero_letras = len(palavra)
print(f"A palavra '{palavra}' tem {numero_letras} letras.")
```

```python
# 7 - Escrever um programa que o usuário digita uma palavra e o programa conta as vogais da palavra.
palavra = input("Digite uma palavra para contar as vogais: ")

vogais = "aeiouAEIOU"
contador_vogais = 0

for letra in palavra:
    if letra in vogais:
        contador_vogais += 1

print(f"A palavra '{palavra}' tem {contador_vogais} vogais.")
```

```python
# 8 - Escrever um programa que o usuário digita uma palavra e o programa imprimi a palavra invertida.
palavra = input("Digite uma palavra para invertê-la: ")

palavra_invertida = palavra[::-1]

print(f"A palavra invertida de '{palavra}' é: '{palavra_invertida}'")
```

```python
# 9 - Escrever um programa que o usuário digita 04 números e o programa imprimi o maior.

print("Digite quatro números:")
numero1 = float(input("Número 1: "))
numero2 = float(input("Número 2: "))
numero3 = float(input("Número 3: "))
numero4 = float(input("Número 4: "))

maior_numero = max(numero1, numero2, numero3, numero4)

print(f"O maior número digitado é: {maior_numero}")
```
